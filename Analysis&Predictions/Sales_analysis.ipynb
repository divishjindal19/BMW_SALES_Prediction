# Little Summary about dataset
Model (11 unique values): Represents various BMW car models such as 5 Series, i8, X1, X3, 7 Series, i3, etc.

Year (15 unique values): Covers car manufacturing years ranging roughly from 2010 to 2024.

Region (6 unique values): Cars are sold across Asia, North America, Middle East, South America, and other global regions.

Color (6 unique values): Includes colors like Red, Blue, Black, White, Silver, etc.

Fuel_Type (4 unique values): Consists of Petrol, Diesel, Hybrid, and Electric.

Transmission (2 unique values): Either Manual or Automatic.

Engine_Size_L (36 unique values): Engine capacities range approximately between 1.0L and 5.0L.

Mileage_KM (Range up to 44,347 km): Indicates driven distance, varying widely across cars.

Price_USD (Range up to $38,246 on average): Represents selling price in USD.

Sales_Volume (Average ≈ 9,845 units): Indicates the number of cars sold for each record.

Sales_Classification (2 categories): High and Low, representing overall sales performance — an ordinal target variable.

```python
import seaborn as sns
plt.figure(figsize=(7,4))
sns.histplot(df["Price_USD"], bins=30, kde=True, color='teal')
plt.title("Price Distribution (USD)")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(7,4))
sns.histplot(df["Engine_Size_L"], bins=20, kde=True, color='orange')
plt.title("Engine Size Distribution (Liters)")
plt.xlabel("Engine Size (L)")
plt.ylabel("Number of Vehicles")
plt.show()

plt.figure(figsize=(7,4))
sns.boxplot(x="Fuel_Type", y="Price_USD", data=df, palette="Set2")
plt.title("Price Distribution by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Price (USD)")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(7,4))
sns.barplot(x="Region", y="Sales_Volume", data=df, estimator='mean', ci=None, palette="mako")
plt.title("Average Sales Volume by Region")
plt.xlabel("Region")
plt.ylabel("Average Sales Volume")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(7,4))
sns.lineplot(x="Year", y="Sales_Volume", data=df, ci=None, marker="o", color="crimson")
plt.title("Sales Trends Over the Years (2010–2024)")
plt.xlabel("Year")
plt.ylabel("Sales Volume")
plt.show()
```

```python
df_fe = df.copy()

#Calculate car age
df_fe['Car_Age'] = 2024- df_fe['Year']
#Price per kilometer
df_fe['Price_per_KM'] = df_fe["Price_USD"]/(df_fe["Mileage_KM"]+1)
#Price per litre of engine
df_fe["Price_per_Litre"] = df_fe["Price_USD"] / (df_fe["Engine_Size_L"] + 0.1)
#Sales-to-price ratio
df_fe["Sales_to_Price_Ration"] = df_fe["Sales_Volume"]/(df_fe["Price_USD"]+1)
#Regional average sales 
region_sales = df_fe.groupby("Region")["Sales_Volume"].transform("mean")
df_fe["Region_Sales_Avg"] = region_sales
# Deviation of price from the global mean
df_fe["Price_Deviation"] = df_fe["Price_USD"] - df_fe["Price_USD"].mean()
# Interaction: engine size * age (older big-engine cars belong to different segments)
df_fe["Engine_Age_Interaction"] = df_fe["Engine_Size_L"] * df_fe["Car_Age"]
# Sales per kilometer (usage efficiency)
df_fe["Sales_per_KM"] = df_fe["Sales_Volume"] / (df_fe["Mileage_KM"] + 1)
# Average sales by fuel type
fuel_sales = df_fe.groupby("Fuel_Type")["Sales_Volume"].transform("mean")
df_fe["Fuel_Sales_Avg"] = fuel_sales
# Average price by color (possible marketing effect)
color_price = df_fe.groupby("Color")["Price_USD"].transform("mean")
df_fe["Color_Price_Avg"] = color_price
```

# Main Conclusion from BMW Sales Dataset
The dataset contains 50,000 fully populated entries across 11 columns, including numerical (Year, Engine_Size_L, Mileage_KM, Price_USD, Sales_Volume) and categorical features (Model, Region, Color, Fuel_Type, Transmission, Sales_Classification).

All columns are complete with no missing values, making it ideal for analysis without additional cleaning.

Visualizations of Year, Model, Sales_Classification, Color, and Fuel_Type indicate that the data is largely consistent, with no major trends, anomalies, or fluctuations over time.

Sales classifications (High, Low) are stable across years and models, suggesting uniform sales performance across the product lineup.

Color and fuel type distributions are balanced or dominant in certain categories but show no extreme variations.

Overall, this dataset provides a stable and consistent overview of BMW sales, which is suitable for categorical analysis, plotting, and potential forecasting, but shows limited variability in patterns across major features.
